```python
from evaluation import Evaluator
from data import FncData

FNC_DATA_PATH = '../../fnc-1/'
```

### Tablica službenih rješenja

```python
competition_data = FncData(
    bodies_csv_path=os.path.join(FNC_DATA_PATH, 'competition_test_bodies.csv'),
    stances_csv_path=os.path.join(FNC_DATA_PATH, 'competition_test_stances.csv'))
competition_data.headlines_bodies_and_stances.head()
```

### Tablice rješenja BERT modela

```python
bert_entailment_with_athene_data = FncData(
    bodies_csv_path=os.path.join(FNC_DATA_PATH, 'competition_test_bodies.csv'),
    stances_csv_path=os.path.join('combined_bert_entailment', \
        'comp_autosave_after_epoch_10_athene.csv'))
bert_entailment_with_athene_data.headlines_bodies_and_stances.head()
```

```python
bert_entailment_with_talos_data = FncData(
    bodies_csv_path=os.path.join(FNC_DATA_PATH, 'competition_test_bodies.csv'),
    stances_csv_path=os.path.join('combined_bert_entailment', \
        'comp_autosave_after_epoch_10_talos.csv'))
bert_entailment_with_talos_data.headlines_bodies_and_stances.head()
```

```python
bert_entailment_with_ucl_data = FncData(
    bodies_csv_path=os.path.join(FNC_DATA_PATH, 'competition_test_bodies.csv'),
    stances_csv_path=os.path.join('combined_bert_entailment', \
        'comp_autosave_after_epoch_10_ucl.csv'))
bert_entailment_with_ucl_data.headlines_bodies_and_stances.head()
```

```python
competition_stances = competition_data.headlines_and_stances['Stance']
evaluator = Evaluator()
```

### Rezultati zajedničkog BERT modela i modela tima Athene

```python
bert_entailment_with_athene_stances = \
    bert_entailment_with_athene_data.headlines_and_stances['Stance']
evaluator.show(competition_stances, bert_entailment_with_athene_stances)
print('Unrelated dodan naknadno:')
print('\tAgree\tDisagree\tDiscuss\n\t{}\t{}\t\t{}'.format(36, 2, 100))
```

```python
score_matrix = evaluator.getScoreMatrix(competition_stances, bert_entailment_with_athene_stances)
score_matrix['Agree'].iloc[3] = 36
score_matrix['Disagree'].iloc[3] = 2
score_matrix['Discuss'].iloc[3] = 100
score_matrix
```

### Rezultati zajedničkog BERT modela i modela tima Talos

```python
bert_entailment_with_talos_stances = \
    bert_entailment_with_talos_data.headlines_and_stances['Stance']
evaluator.show(competition_stances, bert_entailment_with_talos_stances)
print('Unrelated dodan naknadno:')
print('\tAgree\tDisagree\tDiscuss\n\t{}\t{}\t\t{}'.format(41, 0, 197))
```

```python
score_matrix = evaluator.getScoreMatrix(competition_stances, bert_entailment_with_talos_stances)
score_matrix['Agree'].iloc[3] = 41
score_matrix['Disagree'].iloc[3] = 0
score_matrix['Discuss'].iloc[3] = 197
score_matrix
```

### Rezultati zajedničkog BERT modela i modela tima UCL

```python
bert_entailment_with_ucl_stances = \
    bert_entailment_with_ucl_data.headlines_and_stances['Stance']
evaluator.show(competition_stances, bert_entailment_with_ucl_stances)
print('Unrelated dodan naknadno:')
print('\tAgree\tDisagree\tDiscuss\n\t{}\t{}\t\t{}'.format(124, 13, 249))
```

```python
score_matrix = evaluator.getScoreMatrix(competition_stances, bert_entailment_with_ucl_stances)
score_matrix['Agree'].iloc[3] = 124
score_matrix['Disagree'].iloc[3] = 13
score_matrix['Discuss'].iloc[3] = 249
score_matrix
```
